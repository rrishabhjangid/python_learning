```python
def is_valid(s):
    stack = []
    # Map each closing bracket to its opening pair
    pairs = {')': '(', ']': '[', '}': '{'}
    
    for char in s:
        # 1. What do we do if the character is an OPENING bracket?
        if char not in pairs:
            stack.append(char)
            
        # 2. What do we do if the character is a CLOSING bracket?
        else:
            if len(stack)==0:
                return False
            
            top=stack.pop()

            if top!=pairs[char]:
                return False
            
            
    # 3. If we finish the loop and the stack is completely empty, it's valid!
    return len(stack) == 0

is_valid("{[}]")
```
