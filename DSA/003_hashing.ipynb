```python
nums = [2, 7, 11, 15]
target=17

def two_sums(nums,target):
    seen={}

    for i, num in enumerate(nums):
        complement=target-num
        if complement in seen:
            index=seen[complement]
            print(f"Indices: {index},{i}")
        else:
            seen[num]=i

two_sums(nums,target)
```
