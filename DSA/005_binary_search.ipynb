```python
nums = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
target=5

left=0
right=len(nums)-1


while left<=right:
    mid=(left+right)//2
    if nums[mid]==target:
        print(f"Target found ({target})")
        break
    elif target>nums[mid]:
        left=mid+1
    else:
        right=mid-1

else:
    print("Target not found")
```
