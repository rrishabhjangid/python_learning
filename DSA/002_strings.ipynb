```python
sentence = ""
letters = ['h', 'e', 'l', 'l', 'o']

for char in letters:
    sentence += char  

print(sentence)
```

```python
letters = ['h', 'e', 'l', 'l', 'o']
# "".join() allocates exactly enough memory once and builds the string in O(N) time!
sentence = "".join(letters)
print(sentence)
```

```python
word="racecar"

left=0
right=len(word)-1

while left<right:
    if word[left]==word[right]:
        left+=1
        right-=1
    else:
        print("String is not palindrome")
        break

else:
    print("String is a palindrome")
```

```python
#Valid Anagarm
s="listen"

t="silent"

s_dict={}
t_dict={}

for letter in s:
    if letter in s_dict:
        s_dict[letter]=s_dict[letter]+1
    else:
        s_dict[letter]=1

for letter in t:
    if letter in t_dict:
        t_dict[letter]=t_dict[letter]+1
    else:
        t_dict[letter]=1

print(f"Valid Anagram: {s_dict==t_dict}")
```
