Reversing an array without .reverse() or [::-1]

```python
a1=[10,20,30,40,50]
print(a1)
```

```python
left=0
right=len(a1)-1
print(left,right)
```

```python
temp=None
while left<right:
    temp=a1[left]
    a1[left]=a1[right]
    a1[right]=temp
    left+=1
    right-=1
print(a1)
```

Move Zeroes

```python
a1=[0,1,0,3,12]

insert_pos=0

for i in range(len(a1)):
    if a1[i]!=0:
        a1[insert_pos]=a1[i]
        insert_pos+=1

for i in range(insert_pos,len(a1)):
    a1[i]=0

print(a1)
```

```python
# best time to buy and sell a stock
prices = [8, 7, 2, 5, 1, 3]
min_price=float("inf")
profit=0
max_profit=0

for i in range(len(prices)):
    if prices[i]<min_price:
        min_price=prices[i]
    profit=prices[i]-min_price
    if profit>max_profit:
        max_profit=profit

print(f"Profit: {max_profit}")
```
