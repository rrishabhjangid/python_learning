```python
#creating singly linked list
class Node:
    def __init__(self,value):
        self.val=value
        self.next=None
```

```python
head=Node(10)
node2=Node(20)
node3=Node(30)

head.next=node2
node2.next=node3
```

```python
current=head
while current is not None:
    print(current.val)
    current=current.next
```

```python
#adding a node at the start
new_node=Node(5)
new_node.next=head
head=new_node
current=head
while current is not None:
    print(current.val)
    current=current.next
```

```python
#deleting a node at the middle
prev=head
while prev.next is not None:
    if prev.next.val==10:
        prev.next=prev.next.next
    prev=prev.next

current=head
while current is not None:
    print(current.val)
    current=current.next
```

```python
prev=None
current=head
while current is not None:
    temp=current.next
    current.next=prev
    prev=current
    current=temp


current=prev
while current is not None:
    print(current.val)
    current=current.next
```

```python
def reverse_list(head):
    prev = None
    current = head
    
    while current is not None:
        temp = current.next
        current.next = prev
        prev = current
        current = temp
        
    return prev
```
